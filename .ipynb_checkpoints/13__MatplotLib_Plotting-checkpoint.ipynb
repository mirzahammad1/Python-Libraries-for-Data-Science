```python
#import required libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Data from OHIA state, Leading cause of death in the state form year 2012
#cause of death
cause = 'Chronic Diseases','Unintentional Injeries','Alzheriemrs','Infuenza and Pneumonia','Stepsis','Others'
#Percentile
percentile = [62,5,4,2,1,26]
```

```python
#set the pie chart plot properties
#set figure size
plt.figure(figsize=(10,10))
#explode the largest pie in the data set
explode = (0.05,0,0,0,0,0)
#set pie chart properties
plt.pie(percentile,labels=cause,explode=explode,autopct='%1.1f%%',startangle=70)
#set axis equal to draw pie as circle
plt.axis('equal')
#set title of the pie chart
plt.title('OHIA STATE - 2012 : Leading Causes of Death')
#show the plot (pie chart)
plt.show()
```
