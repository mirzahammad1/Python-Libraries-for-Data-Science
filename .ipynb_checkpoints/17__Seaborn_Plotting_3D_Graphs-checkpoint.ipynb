```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ADANIPORTS.csv',parse_dates=True)
```

```python
df.head()
```

```python
df.columns
```

```python
df['H-L'] = df.HIGH - df.LOW
```

```python
df['100MA'] = df.CLOSE
```

```python
#set style darkgird
sns.set_style('darkgrid')
```

```python
#plotting 3D graph
ax = plt.axes(projection='3d')
ax.scatter(df.index,df['H-L'],df['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

```python
import numpy as np
```

```python
z1 = np.linspace(0,10,100)
x1 = np.cos(4*z1)
y1 = np.sin(4*z1)
```

```python
sns.set_style('whitegrid')
ax=plt.axes(projection='3d')
ax.plot3D(x1,y1,z1)
plt.show()
```

```python
def return_z(x,y):
    return 50-(x**2+y**2)
```

```python
sns.set_style('whitegrid')
```

```python
x1,y1=np.linspace(-5,5,50),np.linspace(-5,5,50)
```

```python
x1,y1=np.meshgrid(x1,y1)
z1 = return_z(x1,y1)
```

```python
ax=plt.axes(projection='3d')
ax.plot_surface(x1,y1,z1)
plt.show()
```
