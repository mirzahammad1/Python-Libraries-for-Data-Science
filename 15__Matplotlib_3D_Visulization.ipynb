```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Housing.csv',)
```

```python
df
```

```python
df.head()
```

```python
df['P-A'] = df.price - df.area
df['100MA'] = df.stories
```

```python
ax = plt.axes(projection='3d')
ax.scatter(df['P-A'],df['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('P-A')
ax.set_zlabel('100MA')
ax.show()
```
