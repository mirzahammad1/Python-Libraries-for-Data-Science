```python
# Generating normally distributed data
df = pd.DataFrame({
    'x' : np.random.normal(0,3,10000),
    'y' : np.random.normal(6,4,10000),
    'z' : np.random.normal(-6,6,10000)
})
```

```python
# next plot must be exmine to know that the data is on ame scale
# plotting the data
df.plot.kde()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# it is important to ensure all data is on same scale to improve and algorithms performance
# now we use StandardScaler to keep the value of x,y,z on the same scale. the x,y,z values are displayed on the same scale in the graph
standardscaler = StandardScaler()
data_tf = standardscaler.fit_transform(df)
df = pd.DataFrame(data_tf,columns=['x','y','z'])
df.plot.kde()
```

```python
# now we can see that x y and z are presented on the same scale
```
