```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sc
from sklearn import datasets
from sklearn.preprocessing import OneHotEncoder
from seaborn import load_dataset
```

```python
# data is loaded into pandas data frame data
data = load_dataset('penguins')
```

```python
# Instantiate a One-Hot-Encoder object and assign it to ohe
ohe = OneHotEncoder()
```

```python
# fit and transform the data using the fit_transform() methode
transform = ohe.fit_transform(data[['island']])
```

```python
# it will return the array version to transform data using .toarray() method
print(transform.toarray())
```

```python
# three columns are present in array in the binary form because there are these unique values in the island columns
# print One-Hot encoded categories to know the column labels using .categories_attribute of the encoder
print(ohe.categories_)
```

```python
# add these columns as separate column in the DataFrame
data[ohe.categories_[0]] = transform.toarray()
data
```
